This notebook reads .csv datafiles taken directly from DUKES data on the government website and analyses the data within them to produce plots representing the diversity of the UK's energy supply.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

### 4.1 Natural Gas: Commodity Balances

```python
df = pd.read_csv("DUKES_4.1.csv", thousands=",")
df
```

```python
df = df.set_index("Table 4.1 Natural gas commodity balance (GWh)")
#df
```

```python
x = []
production = []
exports = []
imports = []
tot_supply = []

# Convert the data into lists
for i in np.arange(0,26,1):
    x.append(df.loc['Column1'][i])
    production.append(df.loc['Production'][i])
    imports.append(df.loc['Imports'][i])
    exports.append(df.loc['Exports'][i])
    tot_supply.append(df.loc['Total supply'][i])

# Make a dictionary for the data
supply = {"Production":production, "Imports":imports, "Exports":exports}
colours = ["orange", "skyblue", "mediumseagreen", "yellow", "blue", "chocolate", "plum", "black"]


# Plot Data
fig, ax = plt.subplots()
ax.grid(visible=True, zorder=1)
ax.stackplot(x, supply.values(), labels=supply.keys(), colors=colours, zorder=2)
#ax.plot(x,tot_supply, label="Total Supply", color="black")

# Labels and legend
ax.legend(loc="upper right")
#ax.set_title("Supply of Natural Gas 1998-2023")
ax.set_xlabel("Year")
ax.set_ylabel("Supply [GWh]")
ax.set_xlim(1998,2023)
ax.set_ylim(0,1.6e6)

tick_spacing = 4
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()
```

### 5.1 Energy Supplied Sources

```python
type_e_sup = pd.read_csv("ET_5.1_DEC_24.csv", thousands=",")
type_e_sup = type_e_sup.set_index("Table 5.1 Fuel used in electricity generation and electricity supplied, annual data")
type_e_sup = type_e_sup.replace("[x]",0)
#type_e_sup
```

```python
type_e_sup.iloc[114:125]
#type_e_sup.iloc[129]
float(str(type_e_sup.iloc[114][1]))
```

```python
date = []
coal = []
oil = []
gas = []
nuclear = []
hydro = []
wind = []
solar = []
bio = []
other = []
total = []

for i in np.arange(0,25,1):
    date.append(float(type_e_sup.iloc[86][i+1]))
    coal.append(float(type_e_sup.iloc[114][i+1]))
    oil.append(float(type_e_sup.iloc[115][i+1]))
    gas.append(float(type_e_sup.iloc[116][i+1]))
    nuclear.append(float(type_e_sup.iloc[117][i+1]))
    hydro.append(float(type_e_sup.iloc[118][i+1]))
    wind.append(float(type_e_sup.iloc[119][i+1]))
    solar.append(float(type_e_sup.iloc[123][i+1]))
    bio.append(float(type_e_sup.iloc[124][i+1]))
    other.append(float(type_e_sup.iloc[125][i+1]))
    total.append(float(type_e_sup.iloc[129][i+1]))


# Make a dictionary for the data
supply = {"Natural Gas":gas, "Coal":coal, "Oil":oil, "Nuclear":nuclear, "Hydro":hydro, "Wind":wind, "Solar":solar, "Bioenergy":bio, "Other":other}
colours = ["orange", "skyblue", "mediumseagreen", "yellow", "blue", "chocolate", "plum", "black"]

# Plot Data
fig, ax = plt.subplots(figsize=(12,6))
ax.grid(visible=True, zorder=1)
ax.stackplot(date, supply.values(), labels=supply.keys(), zorder=2)
#ax.plot(date, supply, label="Composition of Energy Supplied", color="black")

# Labels and legend
ax.legend(loc=(0.01,0.93), mode='expand', ncol=9, columnspacing=30, fontsize=10)
#ax.set_title("Composition of Supplied Energy UK 1998-2022", fontsize=16)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Supply [TWh]", fontsize=14)
ax.set_xlim(1998,2022)
ax.set_ylim(0,450)

tick_spacing = 2
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.show()
```

```python
fig.savefig('comp_E_supp.png', dpi=500)
```

### 4.1.1 Natural gas and colliery methane production

```python
ng_sup = pd.read_csv("DUKES_4.1.1.csv", thousands=",")
ng_sup = ng_sup.set_index("Table 4.1.1 Methane production and consumption (GWh) [note 1]")
ng_sup = ng_sup.replace(",","")
#ng_sup
```

```python
f = ng_sup.loc['1998'][3].replace(",","")
f
```

```python
date = []
tot_con = []
domest = []
indust = []
elect = []
other = []
service = []
transport = []

for i in np.arange(0,26,1):
    date.append(1998+i)
    tot_con.append(int(ng_sup.loc[str(1998+i)][3].replace(",","")))
    domest.append(int(ng_sup.loc[str(1998+i)][4].replace(",","")))
    indust.append(int(ng_sup.loc[str(1998+i)][5].replace(",","")))
    elect.append(int(ng_sup.loc[str(1998+i)][6].replace(",","")))
    other.append(int(ng_sup.loc[str(1998+i)][7].replace(",","")))
    service.append(int(ng_sup.loc[str(1998+i)][8].replace(",","")))
    transport.append(int(ng_sup.loc[str(1998+i)][9].replace(",","")))
```

```python
# Plot Data
fig, ax = plt.subplots(figsize=(10,6))
ax.grid(visible=True, zorder=1)

colours = ["orange", "skyblue", "mediumseagreen", "yellow", "blue", "chocolate", "plum", "black"]

ax.plot(date, tot_con, label="Total Consumption", color="black")
ax.plot(date, domest, label="Domestic", color="orange")
ax.plot(date, indust, label="Industry", color="skyblue")
ax.plot(date, elect, label="Electricity", color="mediumseagreen")
ax.plot(date, service, label="Service", color="y")
ax.plot(date, transport, label="Transport", color="blue")
ax.plot(date, other, label="Other", color="plum")

# Labels and legend
ax.legend(loc=(0.575,0.79), ncol=2, columnspacing=1, fontsize=11)
#ax.set_title("Composition of Supplied Energy UK 1998-2022", fontsize=16)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Consumption [GWh]", fontsize=14)
ax.set_xlim(1998,2023)
ax.set_ylim(0,1.3e6)

tick_spacing = 2
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.xticks(fontsize=13, rotation=0)
plt.yticks(fontsize=13, rotation=0)

plt.show()
```

```python
fig.savefig('nat_gas_cons.png', dpi=500)
```
