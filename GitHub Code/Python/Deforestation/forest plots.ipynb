# Comparing forest data from different dates
This notebook contains some plots comparing continental forest coverage between the year 2000 and 2020.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a dictionary of the data
forest_2000 = {'Africa':7.07, 'Asia':13.25, 'Australia':1.48, 'Europe':3.74, 'NA':6.90, 'SA':8.78, 'World':41.23}
forest_2020 = {'Africa':6.74, 'Asia':13.10, 'Australia':1.47, 'Europe':3.80, 'NA':6.75, 'SA':8.34, 'World':40.22}

forest_2020.get('Africa')
```

```python
# Make the data into a 2D array [[2000, 2020],[2000, 2020],...]
#forest_change = [[7.07, 6.74],[13.25, 13.10], [1.48, 1.47],[3.74, 3.80],[6.90, 6.75],[8.78, 8.34],[41.23, 40.22]]
forest_change = [[7.07,13.25,1.48,3.74,6.90,8.78,41.23],[6.74,13.10,1.47,3.80,6.75,8.34,40.22]]
# Make axis for a bar chart
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0,0,1,1])
ax.grid(visible=True)

ax.set_title('Forest Area Compared Between 2000 and 2020', fontsize=19)
ax.set_xlabel('Continent', fontsize=17)
ax.set_ylabel('Forest Area [$km^2$ x $10^6$]', fontsize=17)

# Plot the data in the 2D array
X = np.arange(7)
ax.bar(X, forest_change[0], color = 'g', width = 0.25, label='2000', zorder=10)
ax.bar(X + 0.25, forest_change[1], color = 'b', width = 0.25, label='2020', zorder=10)

# Fix x-axis
ax.set_xticks(X+0.25/2)
ax.set_xticklabels(['Africa','Asia','Australia','Europe','N. America','S. America','World'],fontsize=13)

ax.yaxis.set_tick_params(labelsize=12)

# For each bar in the chart, add a text label.
for bar in ax.patches:
  # The text annotation for each bar should be at its height.
  bar_value = bar.get_height()
  # Format the text with commas to separate thousands. You can do
  # any type of formatting here though.
  text = f'{bar_value:.2f}'
  # This will give the middle of each bar on the x-axis.
  text_x = bar.get_x() + bar.get_width() / 2
  # get_y() is where the bar starts so we add the height to it.
  text_y = bar.get_y() + bar_value
  # If we want the text to be the same color as the bar, we can
  # get the color like so:
  bar_color = bar.get_facecolor()
  # If you want a consistent color, you can just set it as a constant, e.g. #222222
  ax.text(text_x, text_y, text, ha='center', va='bottom', color=bar_color,
          size=12)

ax.legend(loc='upper left', fontsize=14)
```

```python
fig.savefig('Forest Area 2000-2020.png', bbox_inches = 'tight', dpi=400)
```
