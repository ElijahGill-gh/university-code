```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = 'Norm_Mag v Magnet_B.txt'
data = np.genfromtxt(datafile,skip_header=0, delimiter=',', unpack=True)
data
```

```python
datafile3 = 'Norm_Mag v Magnet_B 2.txt'
data3 = np.genfromtxt(datafile3,skip_header=0, delimiter=',', unpack=True)
data3
```

```python
datafile4 = 'Norm_Mag v Magnet_B 3.txt'
data4 = np.genfromtxt(datafile4,skip_header=0, delimiter=',', unpack=True)
data4
```

```python
data = data[3:len(data)-1]
data3 = data3[3:len(data)-1]
data4 = data4[3:len(data)-1]
data
```

```python
len(data4)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.grid(visible=True)
ax.set_xlabel('Magnetic Field, h [T]', fontsize=15)
ax.set_ylabel('Normalised Magnetisation', fontsize=15)
#ax.set_xlim(-1,1)
#ax.set_ylim(-1, 1)
x1 = np.linspace(-5,5,len(data))
x2 = np.linspace(-5,5,len(data3))
x3 = np.linspace(-5,5,len(data4))
ax.set_title('Normalised Magnetisation depending on Magnetic Field', fontsize = 18)
ax.scatter(x1, data, color='red', marker='o', label='Temp = 1K')
ax.scatter(x2, data3, color='blue', marker='o', label='Temp = 10K')
ax.scatter(x3, data4, color='green', marker='o', label='Temp = 100K')
ax.legend()
```

```python
datafile2 = 'Lattice_Spin_Data.txt'
data2 = np.genfromtxt(datafile2, skip_header=0, delimiter=',', unpack=False)
data2
```

```python
len(data2)
```

```python
data2[0:10]
```

```python
data2_list = []
for i in np.arange(0,len(data2)):
    data2_list.append(data2[i][:-1].tolist())
data2_list
```

```python
# Split the list up into the individual lattices
col_len = len(data2_list[0]) # Gives the row/column length, assuming lattice is square
data2_lattice = []
for i in np.arange(0, len(data2_list)):
    if i%col_len == 0:
        data2_lattice.append(data2_list[i:i+col_len])
data2_lattice = np.array(data2_lattice)
data2_lattice[2]
```

```python
fig2 = plt.figure(figsize=(10,10))
ax = fig2.add_subplot()
ax.set_title('Initial Lattice', fontsize = 18)
bar = plt.colorbar(ax.imshow(data2_lattice[0]))
bar.set_label("Spin")
ax.set_xlabel('J = 1, h = 1, mu = 1, temp = 1', fontsize=15)

ax.imshow(data2_lattice[0])
```

```python
fig2 = plt.figure(figsize=(10,10))
ax = fig2.add_subplot()
ax.set_title('Lattice after all evolutions', fontsize = 18)
bar = plt.colorbar(ax.imshow(data2_lattice[0]))
bar.set_label("Spin")
ax.set_xlabel('J = 1, h = 1, mu = 1, temp = 1', fontsize=15)

ax.imshow(data2_lattice[len(data2_lattice)-1])
```
