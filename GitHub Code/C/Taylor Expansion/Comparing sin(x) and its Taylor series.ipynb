```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = 'Sine and Taylor.txt'
data = np.genfromtxt(datafile,skip_header=0, delimiter=',', unpack=True)
data
```

```python
sine = data[:361]
taylor = data[361:722]
```

```python
sine
```

```python
taylor
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.grid(visible=True)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('sin(x)', fontsize=15)
ax.set_xlim(-5,365)
ax.set_ylim(-1.2, 1.2)
x = np.linspace(0,360, 361)
ax.set_title('Comparing sin(x) with its Taylor series', fontsize = 18)
ax.scatter(x, sine, color='red', marker='o', label='sin(x)')
ax.scatter(x, taylor, color='blue', marker='x', label='Taylor')
```

```python
fig2, axs = plt.subplots(1,3, figsize=(20,7))
axs[0].grid(visible=True), axs[1].grid(visible=True), axs[2].grid(visible=True)
axs[2].set_xlabel('x', fontsize=13), axs[2].set_ylabel('f(x)', fontsize=13)
axs[2].set_xlabel('x', fontsize=13), axs[2].set_ylabel('f(x)', fontsize=13)
axs[2].set_xlabel('x', fontsize=13), axs[2].set_ylabel('f(x)', fontsize=13)
axs[0].set_title('f(x) = sin(x)', fontsize = 18)
axs[1].set_title('f(x) = Taylor series of sin(x)', fontsize = 18)
axs[2].set_title('Both functions', fontsize = 18)
axs[0].scatter(x, sine, color='red', marker='o', label='sin(x)')
axs[1].scatter(x, taylor, color='blue', marker='x', label='Taylor')
axs[2].scatter(x, sine, color='red', marker='o', label='sin(x)')
axs[2].scatter(x, taylor, color='blue', marker='x', label='Taylor')
axs[0].legend(loc='upper right', fontsize=16)
axs[1].legend(loc='upper right', fontsize=16)
axs[2].legend(loc='upper right', fontsize=16)
```

```python
fig2.savefig('Comparing sin and Taylor.png', dpi=300)
```
